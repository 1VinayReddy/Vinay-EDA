#### Extracting object type columns into one list and others into another list

```python
# To do that we need to use the help of "df.dtypes"
# But we know that 'df.dtypes' is in series type.
# So, we need to convert series type into dictionary type for easy operations

d1=dict(df.dtypes)
d1
```

```python
l1,l2=[],[]

"""
for i in range(len(d1)):
    if d1[df.columns[i]] == 'object':
        l1.append(df.columns[i])
    else:
        l2.append(df.columns[i])
"""

for i in d1:
    if d1[i] == 'object':
        l1.append(i)
    else:
        l2.append(i)
        
print(l1)
print(l2)
```

```python
categorical=[i for i in d1 if d1[i] == 'object']
numerical=[i for i in d1 if d1[i] != 'object']

print(numerical)
print(categorical)
```

### iloc ---- index location

```python
df.iloc[5:10]              # df.iloc[rows,columns]      
                           # this gives 5 to 10 rows with all the columns
```

```python
df.iloc[:,5:10]              # this gives 5 to 10 columns with all the rows
```

```python
df.iloc[2:20:,5:10]              # this gives intersectional values with 2 to 20 rows and 5 to 10 columns 
```

```python
df.iloc[:,:]              # df.iloc[start:end,start:end]      
                          # this gives all rows with all the columns
```

```python
df.iloc[[100,200,300]] 
```

```python
df.iloc[[100,200,300],[0,1,2,3]] 
```

```python
df.iloc[[100,200,300],[8]]         # if we mention prevailing_wage's index in a list then o/p is in dataframe form
```

```python
df.iloc[[100,200,300],8]         # if we didn't mention prevailing_wage's index in a list then o/p is in series form
```

```python
# df.iloc[[100,200,300],['prevailing_wage']]                 # error -->  .iloc requires numeric indexers
```
