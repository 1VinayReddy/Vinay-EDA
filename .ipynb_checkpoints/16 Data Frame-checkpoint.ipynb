# Create dataframe using list

```python
name=['vinay','jb','bachi']
age=[21,22,21]
```

```python
name,age
```

#### Provide Data

```python
vjb=pd.DataFrame(zip(name,age))           # Creates dataframe
vjb
```

#### Naming Columns

```python
dataa=zip(name,age)
vjb=pd.DataFrame(dataa,columns=["Name","Age"])
vjb
```

```python
data_file=zip(name,age)
column_names=['name','age']
vjb=pd.DataFrame(data_file,columns=column_names)
vjb
```

#### Giving indexes

```python
dataa=zip(name,age)
vjb=pd.DataFrame(dataa,columns=["Name","Age"],index=[1,2,3])
vjb
```

```python
dataa=zip(name,age)
index_order=['a','b','c']
vjb=pd.DataFrame(dataa,columns=["Name","Age"],index=index_order)
vjb
```

#### Adding new columns

```python
area_names=['Habsiguda','Tarnaka','Habsiguda']
vjb['place']=area_names
vjb
```

#### Updating the exesting columns

```python
vjb['Name']
```

```python
vjb['Name']=['Vinay','Jayanand','Bhaskar']
vjb
```

```python
surname=['R.','B.','N.']
vjb['Surnames']=surname
vjb
```

#### Dropping a column

```python
VJB=vjb.drop('Surnames',axis=1)                                # if "inplace=True" is not mentioned then the dataframe will not get updated
VJB                                                            # if axis = 0 -> it means row
                                                               # if axis = 1 -> it means column
# or 

vjb.drop('Surnames',axis=1,inplace=True)
vjb
```

#### Dropping Row

```python
vjb.drop('a',axis=0,inplace=True)                         # deletes 1st row
vjb
```

#### Save the dataframe

```python
vjb.to_csv("G:\\ds&ai\\vjb.csv")                            # give the path -> then name of the file -> then extension type
```
