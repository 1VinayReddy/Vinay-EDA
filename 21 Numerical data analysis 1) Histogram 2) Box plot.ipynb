```python
# Q1. dataframe:-    
# lower_bound           upper_bound             Frequency

l1_lb=[]
l2_ub=[]
l3_f=[]

for index in range(len(n)):
    l1_lb.append(i[index])
    l2_ub.append(i[index+1])
    l3_f.append(f[index])
    
pd.DataFrame(zip(l1_lb,l2_ub,l3_f),columns=['lower_bound','upper_bound','Frequency'])
```
