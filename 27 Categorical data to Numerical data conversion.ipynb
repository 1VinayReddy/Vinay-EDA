### 2. One hot encoder

```python
# this encoder will create no. of columns = no. of unique values present in the data column
# the values are 1 or 0 only
# Unique value's respective column is 1 rest are 0's

# This will create columns that are independent with each other
# In ML independent columns are important for analysis
# if the angle made by the values is 90 degrees then they are independent
#      -> angle between 0( = cos(90) ) and 1( = cos(0) ) is 90 degrees
# " 90 degrees phase shift (or) orthogonal (or) perpendicular -> independent "

# But if the column has 100 unique labels -> 100 new columns (hard to deal)
#     - Data becomes huge
#     - processing time increased
#     - memory consumption is more
#     - This is known as " Curse of Dimentionality "
```
