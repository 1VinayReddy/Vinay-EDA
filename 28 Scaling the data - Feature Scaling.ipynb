# Feature Scaling

### 1. Z-score (Standardization)

```python
# z=(x-mew)/sigma
# 99.7% of data values lies between -3 to 3 range
```

```python
col=visa_df['prevailing_wage']

mean=col.mean()
std=col.std()

z_score=(col-mean)/std
visa_df['prevailing_wage_z']=z_score
visa_df
```

```python
visa_df[['prevailing_wage','prevailing_wage_z']]                        # we need to check whether the data is transformed correctly or not
```

```python
visa_df['prevailing_wage'].max(),visa_df['prevailing_wage_z'].max()        # if the id's of max are same then our calculaion is correct
```

```python
visa_df['prevailing_wage'].min(),visa_df['prevailing_wage_z'].min()
```

```python
visa_df['prevailing_wage'].idxmax(),visa_df['prevailing_wage_z'].idxmax()          # idxmax will give you the id of max value
```

```python
visa_df['prevailing_wage'].idxmin(),visa_df['prevailing_wage_z'].idxmin()
```

### 2. min-max scalar (Normalization)

```python
# minmax = (x - x_min)/(x_max - x_min)
# range from -1 to 1
```

```python
path="G:\ds&ai\Visadataset.csv"
visa_df=pd.read_csv(path)
```

```python
col=visa_df['prevailing_wage']

min_max=(col-col.min())/(col.max()-col.min())
visa_df['prevailing_wage_minmax']=min_max
```

```python
visa_df[['prevailing_wage','prevailing_wage_minmax']]
```

```python
visa_df['prevailing_wage'].max(),visa_df['prevailing_wage_minmax'].max()
```

```python
visa_df['prevailing_wage'].idxmax(),visa_df['prevailing_wage_minmax'].idxmax()
```

```python
visa_df['prevailing_wage'].min(),visa_df['prevailing_wage_minmax'].min()
```

```python
visa_df['prevailing_wage'].idxmin(),visa_df['prevailing_wage_minmax'].idxmin()
```
