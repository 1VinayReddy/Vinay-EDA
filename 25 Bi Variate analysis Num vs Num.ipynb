```python
num_cols=visa_df.select_dtypes(exclude='object')
num_cols.columns
```

```python
# value close to 1 -> highly related or dependent (directly related)
# value close to 0 -> almost independent or not related
# value close to -1 -> highly related or dependent (inversely related)
```

```python
path="G:\ds&ai\winequality_red.csv"
wine_df=pd.read_csv(path)
wine_df
```

```python
wine_df.select_dtypes(exclude='object')
```

```python
wine_df.corr(numeric_only=True)
```

```python
wine_df.columns
```

```python
# fixed acidity vs citric acid has correlation value=0.67...
# directly related

col1=wine_df['fixed acidity']
col2=wine_df['citric acid']

plt.scatter(col1,col2)
plt.show()
```

```python
# fixed acidity vs pH has correlation value= -0.68...
# inversely related

col1=wine_df['fixed acidity']
col2=wine_df['pH']

plt.scatter(col1,col2)
plt.show()
```

```python
# with df.corr() the only problem is when the no. of columns are more
# if we have 1000 columns what can we do???

# so we use heatmap
```

## Heatmap

```python
# It is a visualization of correlation matrix
# In seaborn package
# Change in color represents change in value of correlation matrix
```

```python
corr_visa=visa_df.corr(numeric_only=True)
corr_visa
```

```python
sns.heatmap(corr_visa)         # sns.heatmap(<correlation matrix>)
```

```python
sns.heatmap(corr_visa,annot=True)            # annot means annotation 
plt.show()
```

```python
corr_wine=wine_df.corr(numeric_only=True)
corr_wine
```

```python
plt.figure(figsize=(12,10))
sns.heatmap(corr_wine,annot=True)
plt.show()
```
