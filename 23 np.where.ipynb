### np.where

```python
dict1={'name':['a','b','c','d'],'num':[1,2,3,4]}
df=pd.DataFrame(dict1)
df
```

```python
# Q.  replace num which has value > 2 with 100
#     otherwise dont change

# Ans empty list -> for loop --> if-else --> append --> df['']=list
```

```python
# instead of that use np.where
# np.where(<condition>,<condition true o/p>,<condition false o/p>)
# np.where will replace the value of the condition
#         if the condition is true replaces present value with true o/p
#         if the condition is false replaces present value with false o/p

l=[]

condition=df['num']>2
tru=100
fals=df['num']

l=np.where(condition,tru,fals)   
df['num']=l
df
```
